```python
def calculate_correlation_matrix(data):
    n = len(data[0])
    M = numpy.zeros((n,n))
    for k in range(len(data)):
        if(k/100 == 1):
            print "."
        for i in range(n):
            for j in range(n):
                if i!= j:
                    if(data[k][i] == data[k][j]):
                        M[i][j] += 1
                    else:
                        M[i][j] -= 1
    M *= 1.0/len(data)
    return M
```

```python
G = StructureLearn.generateRandomBayesNet_withPriors(25, 80, 5)
```

```python
def genGraph(M,threshold):
    G = nx.Graph()
    for i in range(len(M[0])):
        G.add_node(i)
    for i in range(len(M[0])):
        for j in range(len(M[0])):
            if(M[i][j] > threshold):
                G.add_edge(i,j, weight = M[i][j])
    return G
```

```python
data = StructureLearn.generateSamples_alt(G, 5000)
```

```python
M2 = calculate_correlation_matrix(data)
```

```python
def Mat_deconvolve(M):
    i = numpy.identity(len(M[0]))
    L = numpy.linalg.matrix_power(i - M, -1)
    A2 = numpy.matmul(M, L)
    return A2
```

```python
def eigen_deconv(M):
    n = len(M[0])
    w, v = numpy.linalg.eig(M)
    dirs = numpy.zeros(n, dtype = complex)
    for i in range(len(w)):
        inv = 1.0/(1 + w[i])
        d = w[i]*inv
        dirs[i] = d
    diag = numpy.zeros((n,n), dtype= complex)
    for i in range(len(dirs)):
        diag[i][i] = dirs[i]
    A_prime = numpy.zeros((n,n), dtype = complex)
    numpy.matmul(v, diag, A_prime)
    A_final = numpy.zeros((n,n), dtype = complex)
    numpy.matmul(A_prime, numpy.linalg.matrix_power(v, -1), A_final)
    return A_final
```
